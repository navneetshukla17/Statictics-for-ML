## 1. Working with Categorical Data / Qualitative Data

```python
data['Embarked'].value_counts().plot(kind='bar')
```

```python
data['Pclass'].value_counts().plot(kind='pie', autopct='%0.f')
```

# Cumlative Frequency of Pclass

```python
a = data['Pclass'].value_counts()
```

```python
# formula
b = ((a / a.sum()) * 100).values
b
```

```python
c = []
cf = 0

for i in b:
    cf += i
    c.append(cf)
c
```

```python
# Plotting Cumlative Frequencies
plt.plot(c, marker='o')

plt.title('Cumlative Frequency of `Pclass` Cloumn')
plt.xlabel('Pclass')
plt.ylabel('Count')

plt.grid(True)
plt.show()
```
