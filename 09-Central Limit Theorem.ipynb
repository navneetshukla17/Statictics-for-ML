```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
```

```python
delivery = pd.read_csv('deliveries.csv')
```

```python
runs = delivery.groupby('batsman').sum()['batsman_runs']
```

```python
runs = runs.values
```

```python
sns.distplot(runs)
```

```python
# n = 50/461, batsman taken per sample
# m = 1000, number of times sampling done
```

```python
np.random.choice(runs, size=10)
```

```python
m,n = 1000, 50

new = []

for i in range(m):
    x = np.random.choice(runs, size=n)
    new.append(x.mean())
```

```python
len(new)
```

```python
type(new)
```

```python
# converting list into an numpy array
new = np.array(new)
```

```python
type(new)
```

```python
sns.distplot(new)
```

```python
runs.mean()
```

```python
new.mean() # new's mean is very close to run's mean, hence Central Limit Theorem is Correct !
```

```python
runs.var() / 50
```

```python
new.var()
```
