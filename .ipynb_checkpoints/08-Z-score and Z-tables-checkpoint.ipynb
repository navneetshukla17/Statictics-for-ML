## 1. Standardization

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
```

```python
from scipy.stats import norm
# Generate 1000 normal random integers with specified mean and std.
# loc [mean], scale [standard deviation]
draw = norm.ppf(np.random.random(10000), loc=1000, scale=200).astype(int)
sns.distplot(draw)
```

```python
z = []

for i in draw:
    z.append((i-1000)/200) # every item - mean / std
```

```python
a = np.array(z)
```

```python
print(a.mean())
print(np.round(a.mean()))
```

```python
print(a.std())
print(np.round(a.std()))
```

```python
sns.distplot(a)
```

## 2. Calculating Z-score

Problem statement: Imagine a group of 300 students who took a test Rohit scored 700 out of 1000. The average score was 500 and the standard deviation was 120. Assuming that the test scores followed normal distribution, find out how well Rohit performed in the test in comparision to his peers.  

```python
# Step 1- calculate z-score 
# formula: z-score = (xi - mean) / std

z = (700 - 500) / 120
print(z)
```

```python
# Step 2- Lookup the Z-table [using z-table.com]
a = 0.9515
```

```python
# Step 3- Multiply by 100
a * 100
```

### Conclusion:

Rohit performed better than 95% of his peers.
