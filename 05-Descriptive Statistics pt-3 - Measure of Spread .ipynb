# Measures of spread

Measures of spread gives us an idea of how much the data is spread OR varying. Some important measures of spreads are:

1. Range
2. Interquatile Range (IQR)
3. Variance
4. Standard Deviation

# 1. Range

The range describe the difference between the largest point and the smallest point in the data.

```python
range_ = np.max(iris_setosa['PL']) - np.min(iris_setosa['PL'])
print(f'Range is: {range_}')
```

# 2. Percentiles

A percentile is a measure used in statistics indicating the value below which a given percentage of observations in a group of observations falls. For example, the 20th percentile is the value below which 20% of the observations may be found.

```python
mylist = [1, 2, 3, 4, 5, 6, 7, 8, 9]

print(np.percentile(mylist, 25))
print(np.percentile(mylist, 90))
```

```python
np.percentile(iris_setosa['PL'], 95) # 95% of the value in iris_setosa column is roughly less than 1.7
```

# 3. Quartiles
Quartiles are the value that divide a list of numbers into quarters:

- Put the list of numbers in order (sort the values)
- Then cut the list into four equal parts
- The quartiles are at the 'cuts'

```python
print(np.percentile(a=iris_setosa['PL'], q=25))
print(np.percentile(a=iris_setosa['PL'], q=50))
print(np.percentile(a=iris_setosa['PL'], q=75))
print(np.percentile(a=iris_setosa['PL'], q=100))
```

# 4. Interquartile Range

The interquartile range (IQR) is a measure of statistical despersion between upper (75th) and lower (25th) quartiles

```python
iqr = np.percentile(a=iris_setosa['PL'], q=75) - np.percentile(a=iris_setosa['PL'], q=25)
print(f'The IQR of Petal Length in iris_setosa column is: {iqr}')
```

# 5. Mean Absolute Deviation
The Absolute deviation from mean, also called `Mean Absolute Deviation (MAD)`, describes the variation in the dataset, In sense it tells that the average distance of each data point in the set is calculated as:



### Example:

2, 5, 1, 6, 7, 3, 11

Mean = 2+5+1+6+7+3+11 / 7 (total no. of observation) = 35 / 7 = 5

The mean is `5`

Mean Absolute Deviation = (|2-5| + |5-5| + |1-5| + |6-5| + |7-5| + |3-5| + |11-5|) / 7 = 3+0+4+1+2+2+6 = 18 / 7 = 2.5

The Mean Absolute Deviation is `2.5`

# 6. Variance 

Variance measures how far are the data points spread out from the mean.

The variance is computed by finding the difference between every data point and the mean, squaring them, summing them up and then again taking average of those numbers like this:



The squares are used during the calculation because they weight outliers more heavily than points that are near to the mean. This prevents that differences above the mean neutralize those below the mean.

### Example:

2, 5, 1, 6, 7, 3, 11

The mean is 5

Variance = ((2-5)^2 + (5-5)^2 + (1-5)^2 + (6-5)^2 + (7-5)^2 + (3-5)^2 + (11-5)^2) / 7

9+0+16+1+4+4+36+ = 70/7 = 10

```python
np.var(iris_setosa['PL'])
```

# The problem with `Variance` !

Because of the squaring part in the calculation it shows the spread in terms of `square`, it basicly changes any `unit` to `unit-square` and that's not the best measurment of our data.

# 7. Variance Vs Median Absolute Deviation (MAD)

Variance is easy to work with, In comparison to MAD as it works on squaring (smooth function) where as, the deviation uses modulus (non-smooth) function.

Mathematically speaking smooth functions are easier to work with as compared to non-smooth functions.



# 8. Standard Deviation

The sqaure root of Variance is called as the `Standard Deviation`. It is calculated as:


### Example:
2, 5, 1, 6, 7, 3, 11

The mean is 5

Variance = ((2-5)^2 + (5-5)^2 + (1-5)^2 + (6-5)^2 + (7-5)^2 + (3-5)^2 + (11-5)^2) / 7

9+0+16+1+4+4+36+ = 70/7 = 10

Variance is 10

Standard Deviation is square root of Variance = sqrt(10) = 3.6

```python
np.std(iris_setosa['PL'])
```

# 9. Median Absolute Deviation 

MAD is the median of all the number obtained from substracting and calculating the absolute value of every observation with median

## Example:
1, 3, 5, 6, 7, 8, 9

Median of any dataset is the middle number of that dataset.

Here, Median is 6

|1-6|, |3-6|, |5-6|, |6-6|, |7-6|, |8-6|, |9-6|

5, 3, 1, 0, 1, 2, 3 (sort this values)

0, 1, 1, 2, 3, 3, 5, (and the median of this dataset after sorting is called `Median Absolute Deviation (MAD)`)

Therefore, Median Absolute Deviation is 2.

```python
from statsmodels import robust
```

```python
robust.mad(a=np.array([1, 3, 5, 6, 7, 8, 9])) # expected output was 2 but, it gave 2.96..
```

Just like `Mean` Vs `Median`, median absolute deviation is free from the effects of outliers from which Standard Deviation can suffer !, So when we have Outliers in your data (Use MAD, median absolute deviation) and If your data has no outliers you must go with `Standard Deviation`
